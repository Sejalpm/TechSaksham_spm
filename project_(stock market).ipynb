```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go

def fetch_stock_data(ticker, start_date, end_date):
    stock = yf.download(ticker, start=start_date, end=end_date)
    return stock

def plot_stock_trends(stock_data, ticker):
    plt.figure(figsize=(10, 5))
    plt.plot(stock_data['Close'], label=f'{ticker} Closing Price')
    plt.title(f'{ticker} Stock Price Trend')
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.legend()
    plt.show()

def interactive_stock_chart(stock_data, ticker):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=stock_data.index, y=stock_data['Close'], mode='lines', name='Closing Price'))
    fig.update_layout(title=f'{ticker} Stock Price Trend', xaxis_title='Date', yaxis_title='Price')
    fig.show()

if __name__ == "__main__":
    ticker_symbol = "AAPL"  # Replaced input() with a predefined ticker for sandbox compatibility
    start_date = "2023-01-01"
    end_date = "2024-01-01"
    
    stock_data = fetch_stock_data(ticker_symbol, start_date, end_date)
    print(stock_data.head())
    
    plot_stock_trends(stock_data, ticker_symbol)
    interactive_stock_chart(stock_data, ticker_symbol)
```
