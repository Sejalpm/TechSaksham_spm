```python
import matplotlib.pyplot as plt

# Define data
x1_axis = [1, 2, 3, 4, 5]
y1_axis = [10, 20, 25, 30, 50]

x2_axis = [1, 2, 3, 4, 5]
y2_axis = [5, 15, 20, 25, 40]

# First subplot
plt.subplot(1, 2, 1)
plt.plot(x1_axis, y1_axis, 'r')  # 'r' for red
plt.title("Graph 1")

# Second subplot
plt.subplot(1, 2, 2)
plt.plot(x2_axis, y2_axis, 'g')  # 'g' for green
plt.title("Graph 2")

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.title('exponential')
plt.plot(x1_axis,y1_axis,'blue')

plt.subplot(2,1,2)
plt.title('linear')
plt.plot(x1_axis,y1_axis,'black')

plt.tight_layout()
plt.show()
```
